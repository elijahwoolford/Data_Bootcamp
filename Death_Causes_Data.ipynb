```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("/Users/elijahwoolford/Desktop/python /NCHS_-_Leading_Causes_of_Death__United_States.csv")
df.head()
```

```python
df2 = df.loc[lambda x: df["Cause Name"] == "Homicide", :]
df2.head()
```

```python
df3 = df2.groupby("Year").mean()
df3.head()
```

```python
#Plot 1

ax = df3["Deaths"].plot(kind="bar", figsize=(10,7), legend=True, fontsize=12)
ax.set_xlabel("Year")
ax.set_ylabel("Death")
plt.title("Homicide by Year")
plt.tight_layout()
plt.show()
```

Pros: This plot clearly shows the relationship of deaths per year
Cons: Doesn't accurately convey the trend over time

```python
#Plot 2

x = df3.reset_index()
g = sns.lmplot(x="Year", y="Deaths", truncate=True, size=5, data=x)
g
```

Pros: This plot is a good indicator of the trend over time
Cons: Doesn't accuartely depict relationship
